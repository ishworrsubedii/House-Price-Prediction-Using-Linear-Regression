# Linear Regression

```python
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
x=house_data[['bedrooms','bathrooms','sqft_living']]
y=house_data['price']
x_features=x.columns
```

```python
x_train=np.array(x)
y_train=np.array(y).reshape(-1,1)
```

```python
fig,ax=plt.subplots(1,3,figsize=(12,12),sharey=True)
for i in range(len(ax)):
    
    ax[i].scatter(x_train[:,i],y_train,label='Actual data')
    ax[i].set_xlabel(x_features[i])
ax[0].set_ylabel('Price')
plt.show()
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
modelprediction=lr.predict(x_train)

w=lr.coef_
b=lr.intercept_
print(f'w:{w}b:{b}')
```

```python
fig,ax=plt.subplots(1,3,figsize=(12,12),sharey=True)
for i in range(len(ax)):
    
    ax[i].scatter(x_train[:,i],y_train,label='Actual data')
    ax[i].set_xlabel(x_features[i])
    ax[i].scatter(x_train[:,i],modelprediction,label='Model Prediction',c='r')
ax[0].set_ylabel('Price')
plt.show()
```

# Learning Curve

```python
def cost(x,y,w,b):
  
    m =3
    cost = 0.0
    

    
    for i in range(m):
        xii = x[i]
        f_wb_i = np.dot(xii, w) + b           #(n,)(n,) = scalar (see np.dot)
        cost = cost + (f_wb_i - y[i])**2       #scalar
    cost = cost / (2 * m)                      #scalar    
    return cost
    
def learning_rate(x,y,w,b):
    j_history=[]

    for i in range(iteration):
        j_history.append( cost(x, y, w, b))
    return j_history


hist=learning_rate(x_train,y_train,w,b)
```

```python
plt.plot(np.arange(iteration),hist)
plt.axis([0,iteration,0,None])
plt.xlabel('Iterations')
plt.show()
```

```python
x_train=np.array(x)
y_train=np.array(y).reshape(-1,1)


from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import numpy as np

X_train, X_test, y_train, y_test = train_test_split(x_train, y_train, test_size=0.3, random_state=42)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

model = Ridge(alpha=0.1)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error on test set:", mse)

coeffs = model.coef_
intercept = model.intercept_
print("Coefficients:", coeffs)
print("Intercept:", intercept)
```
